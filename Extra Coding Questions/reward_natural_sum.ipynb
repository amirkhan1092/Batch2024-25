### Problem Statement:

A company is distributing rewards to employees based on their ranks. The ranks are numbered sequentially from 1 to \( N \), where \( N \) is the total number of employees.  

The reward system works such that:
- The first-ranked employee receives 1 reward point.
- The second-ranked employee receives 2 reward points.
- The third-ranked employee receives 3 reward points.
- And so on, up to the \( N \)-th employee who receives \( N \) reward points.

Write a program to calculate the **total reward points** distributed to all employees.  

---

### Input Format:  
- A single integer \( N \), representing the total number of employees.  

---

### Output Format:  
- A single integer representing the total reward points distributed to all employees.  

---

### Constraints:  
- \( 1 \leq N \leq 10^9 \)  

---

### Example:  

#### Example 1:  
**Input:**  
```plaintext
5
```  

**Output:**  
```plaintext
15
```  

**Explanation:**  
The reward points are:  
\[
1 + 2 + 3 + 4 + 5 = 15
\]  

---

#### Example 2:  
**Input:**  
```plaintext
10
```  

**Output:**  
```plaintext
55
```  

**Explanation:**  
The reward points are:  
\[
1 + 2 + 3 + \dots + 10 = 55
\]  

---

### Constraints on the Problem:  
1. Avoid using loops for computation, as \( N \) can be very large.  
2. Use an efficient mathematical approach to solve the problem.  

---

### Solution Code:  
```python
def calculate_rewards(n):
    # Use the formula to calculate the sum of the first N integers
    return n * (n + 1) // 2

# Input Handling
if __name__ == "__main__":
    n = int(input())
    print(calculate_rewards(n))
```

---

### Complexity Analysis:  

1. **Time Complexity**:  
   - \( O(1) \): The formula is computed in constant time.  

2. **Space Complexity**:  
   - \( O(1) \): No additional data structures are used.
