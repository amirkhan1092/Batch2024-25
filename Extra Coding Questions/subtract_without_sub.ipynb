### Problem Statement:  
Write a program to subtract two integers without using the subtraction (`-`) operator. You can use only addition, bitwise, or logical operators to solve the problem.

---

### Input Format:  
- The first line contains two integers, `a` and `b`, separated by a space.  

---

### Output Format:  
- Print the result of `a - b`.  

---

### Constraints:  
- \(-10^9 \leq a, b \leq 10^9\)  
- You must not use the subtraction operator (`-`) in your solution.

---

### Examples:  

#### Example 1:  
**Input:**  
```plaintext
10 4
```  

**Output:**  
```plaintext
6
```

---

#### Example 2:  
**Input:**  
```plaintext
5 8
```  

**Output:**  
```plaintext
-3
```

---

### Hint:  
To subtract two numbers without using the subtraction operator:
1. Use the relation `a - b = a + (-b)`.
2. Compute the negation of `b` using bitwise NOT and addition.

---

### Expected Solution in Java:  
```java
public class Main {
    public static void main(String[] args) {
        java.util.Scanner scanner = new java.util.Scanner(System.in);
        int a = scanner.nextInt();
        int b = scanner.nextInt();
        scanner.close();

        // Subtract b from a without using the '-' operator
        int result = add(a, negate(b));
        System.out.println(result);
    }

    // Method to negate a number without using '-'
    private static int negate(int x) {
        return ~x + 1; // Use bitwise NOT and add 1 (two's complement)
    }

    // Method to add two numbers
    private static int add(int x, int y) {
        while (y != 0) {
            int carry = x & y; // Find carry
            x = x ^ y;         // Perform addition without carry
            y = carry << 1;    // Shift carry to the left
        }
        return x;
    }
}
```

---

### Key Techniques:
1. **Two's Complement**: Negate a number using `~x + 1` to compute `-x`.
2. **Bitwise Operations**: Perform addition using XOR (`^`) and carry propagation with AND (`&`) and bit shifts (`<<`).
3. **Mathematical Properties**: Leverage `a - b = a + (-b)` to achieve subtraction.
